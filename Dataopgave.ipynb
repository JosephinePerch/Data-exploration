### Datasæt om køn og alder
**Vi kigger på hvilke grupper, der er i datasættet.**

```python
print(df2.Group.unique())
print('Antal grupper:', len(df2.Group.unique()))
```

Der er 12 grupper med forskellig alder og køn. To af disse er en samlet gruppe for hhv. mænd og kvinder.  

**Vi kigger nu på emnerne i datasættet om køn og alder:**

```python
print(df2['Topic Label'].unique())
print('Antal Topic Labels: ', len(df2['Topic Label'].unique()))
print('Antal Question Labels: ', len(df2['Question Label'].unique()))
```

**Vi finder korrelationen mellem spørgsmålenes scorer.**  
Datasættet pivoteres, så der dannes en række for hver gruppe af køn og alder, hvor scoren angives for hvert af de 109 Question Labels. De to rækker for hhv. samlet kvinder og samlet mænd slettes. 

Ud fra det nye datasæt findes korrelationen mellem variablene. Det giver os en korrelations matrix, som vi kan plotte.  
(NB: Det vil give stærkere information at udregne korrelations koefficienten mellem enkelte observationer og ikke mellem grupperede observationer som her. En del af signalet forsvinder, når middelværdien over observationerne i grupperne beregnes.)

```python
df2pivot = pd.pivot_table(df2, values = 'Score', index = 'Group', columns = ['Question Label'])
df2pivot.drop(df2pivot.index[[0,6]])
corr = df2pivot.corr(method ='pearson') 

ax = sns.heatmap(
    corr, 
    vmin = -1, vmax = 1, center = 0,
    cmap = sns.diverging_palette(20, 220, n = 200),
    square = True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 45,
    horizontalalignment = 'right');
```

Matricen er så stor, at den her er umulig at aflæse. 

**For at gøre det mere overskueligt kigger vi kun på korrelationen mellem de første 10 spørgsmål:**

```python
corr2 = corr.iloc[0:10, 0:10]

ax = sns.heatmap(
    corr2, 
    vmin = -1, vmax = 1, center = 0,
    cmap = sns.diverging_palette(20, 220, n = 200),
    square = True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 45,
    horizontalalignment = 'right');
```

Eksempelvis kan vi se at grupper, der har scoret højt på "Arbejde tager energi fra privatliv" også har scoret højt på "Arbejdsrelateret stress" og omvendt.  
Tilsvarende er der fx negativ korrelation mellem "Arbejdet giver selvtillid og arbejdsglæde" og "Andet end fast dagarbejde". Her er det vigtigt for fortolkningen hvad en lav eller en høj score er et udtryk for. En lav score for "Andet end fast dagarbejde" betyder fx at man har fast dagarbejde (kl. 06-18).  
Et andet eksempel er en korrelation tæt på 0 mellem "Arbejdsevne" og "Arbejde tager tid fra privatliv".


**Vi undersøger fordelingen af *Uoverskuelighed og stress* over aldersgrupper og køn.**

```python
df2new = df2.loc[df2['Topic Label'] == 'Uoverskuelighed og stress']
df2new = df2new.loc[df2new['Akse'] == 'Score']
df2new = df2new.iloc[::-1] #Reverse order of rows to get the youngest first
#df2new.head(24)
```

```python
sns.catplot(x = 'Group', y = 'Score', hue = 'Question Label', kind = 'bar', data = df2new)
plt.xticks(rotation = 45, horizontalalignment = 'right')
plt.title('Uoverskuelighed og stress');
```

Det kan ses på plottet, at kvinder generelt scorer højere på stress og uoverskuelighed end mænd. For kvinder er stress-scoren højere end scoren for uoverskuelighed, og for mænd er det omvendt.

For begge køn scorer personer i alderen 25 til 34 år højest på både stress og uoverskuelighed.

**Vi undersøger fordelingen af *Arbejdsevne* over aldersgrupper og køn.**

```python
df2new2 = df2.loc[(df2['Question Label'] == 'Psykisk arbejdsevne') | (df2['Question Label'] == 'Fysisk arbejdsevne')]
df2new2 = df2new2.iloc[::-1] #Reverse order of rows 
#df2new.head(24)
```

```python
sns.catplot(x = 'Group', y = 'Score', hue = 'Question Label', kind = 'bar', data = df2new2)
plt.xticks(rotation = 45, horizontalalignment = 'right')
plt.title('Arbejdsevne');
```

Her er en score på 1 lig med dårlig arbejdsevne, og 5 er fremragende.  
Der er en generel tendens til, at den fysiske arbejdsevne er højere end den psykiske. Generelt ser det ud til, at forskellen på dem bliver mindre med alderen, hvor den fysiske falder.
